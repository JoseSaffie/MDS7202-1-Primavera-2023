**2. La funci√≥n `describir(datos)` que generalice la funci√≥n `generar_descripcion` para recibir una lista de frutas y devolver una lista con las descripciones de cada una:**

Por ejemplo, `describir([limon, naranja]])` deber√° devolver una lista de la forma `[descripcion_limon, descripcion_naranja]`

````python
def describir(datos):
    # Documentaci√≥n Aqu√≠
    # Recibe una lista de frutas, para luego entregar una lista con la descripcion de cada una
    # Creamos una lista vacia
    descripciones = []
    # Llamamos la funcion generar_descripci√≥n en un loop de for para que recorra la lista de frutas
    for i in datos:
        des = generar_descripcion(i)
        descripciones.append(des)

    # Retorna una lista de frutas en formato de string
    return descripciones

describir([limon,naranja])
````

**3. La funci√≥n `filtrar_por_pepa(datos, tiene_pepas)` que dado un tipo de fruta y un booleano, retorne las frutas que tienen o no tienen pepas (seg√∫n el valor de `tiene_pepas`).**
    

Por ejemplo, `filtrar_por_pepa(datos, True)` deber√° retornar una lista con los diccionarios de: naranja, limon, frutilla y frambuesa.

````python
def filtrar_por_pepa(datos, tiene_pepas):
    # Documentaci√≥n Aqu√≠
    # Filtramos el diccionario usando comprension, para obtener un nuevo diccionario solo con el booleano que se quiere
    
    # C√≥digo Aqu√≠
    frutas = [fruta for fruta in datos if fruta["pepas"] == tiene_pepas]

    # Luego creamos una lista vacia para llenara en un for con los nombre de las frutas
    filtrados = [] 
    for fruta in frutas:
        filtrados.append(fruta["nombre"])
   
   # Finalmente retornamos una lista de frutas en string.
    return filtrados

filtrar_por_pepa(datos, True)
````

**4. La funci√≥n `conteo_colores(datos)` que cree un diccionario que haga un conteo los colores de las frutas.**

Para estos datos, la funci√≥n deber√≠a retornar el siguiente diccionario: 
    
```python
{"amarillo": 3, "naranjo": 1, "rojo": 2}
```

````python
def conteo_colores(datos):
    # Documentaci√≥n Aqu√≠
    # Primero definimos un diccionario vacio
    # Codigo aqu√≠
    conteo = {}
    #Entonces nos tenemos que poner en 2 casos, cuando un color todavian no esta en el diccionario y cuando un color ya esta,
    # recorremos el diccionario grande con un for

    for fruta in datos:
        # Recorremos el diccionario extrayendo los colores
        if "color" in fruta:
            color = fruta["color"]
            # Caso que ya existe
            if color in color in conteo:
                conteo[color] = conteo[color] + 1
            # Caso que no existe
            else: 
                conteo[color] = 1 
    # Finalmente retornamos un diccionario con el conteo de colores        
    return conteo

conteo_colores(datos)
````

**5. La funci√≥n `obtener_tipos` que devuelva una lista con los tipos √∫nicos de fruta usando `set()`**.

Por ejemplo, `obtener_tipos([limon, naranja, platano])` deber√≠a devolver `["citrica", "tropical"]`

````python
def obtener_tipos(datos):
    # Documentaci√≥n Aqu√≠}
    # El input son los nombres de las frutas y se ponen a modo de lista, retorna una lista de strings
    # Creamos la lista vacia
    tipos = []

    # Recorremos todas la frutas y obtenemos que tipo de frutas son
    for fruta in datos:
        tipos.append(fruta["tipo"])
    
    # Utilizamos set para eliminar duplicados
    tipos = set(tipos)

    # retorna una lista de strings
    return tipos

obtener_tipos([limon])
````

    - Verifique que ninguna fruta tenga pepas. En el caso que haya alguna, imprimir (con `print`) el mensaje de advertencia `'Alerta ‚ö†Ô∏è: El jugo puede contener restos de pepas.'`. Puede usar la funci√≥n definida en la secci√≥n anterior.

````python
class Jugera(Electrodomestico):
    
    def __init__(self):
      super().__init__()
      self.bandeja = []
              
    def agregar_ingrediente(self, nueva_fruta):
        self.bandeja.append(nueva_fruta)
            
    def listar_ingredientes(self):
        if not self.bandeja:
            print('Bandeja vacia')
        else:
            separador = ", "
            cadena_resultante = separador.join([fruta["nombre"] for fruta in self.bandeja])
            print(f'Ingredientes de la bandeja: {cadena_resultante}.')
            
    def preparar_jugo(self):
        self.esta_enchufado()
        if not self.bandeja:
            raise Exception('Error ‚ùå: La bandeja no tiene ingredientes.')

        frutas_con_pepas = filtrar_por_pepa(self.bandeja, True)
        if frutas_con_pepas:
            print('Alerta ‚ö†Ô∏è: El jugo puede contener restos de pepas.')
                        
        separador = ", "
        cadena_resultante = separador.join([fruta["nombre"] for fruta in self.bandeja])
        mensaje = f'Jugo de {cadena_resultante} listo. üèñÔ∏èü•§ Que lo disfrutes!!! ü•§üèñÔ∏è.'
        self.bandeja.clear()
        return mensaje
````

### Clase Jugera + Properties  [1.0 puntos]

Implementar `bandeja` usando una `property` que permita setear una `nueva_bandeja` como `bandeja` seg√∫n las siguientes condiciones:


- Compruebe que `nueva_bandeja` sea una lista. En caso contrario, levante una excepci√≥n.
- No permita agregar m√°s de 3 ingredientes a la bandeja a la vez. Si se entregan m√°s de 3 frutas, se levante una excepci√≥n.
- Se compruebe que todos los elementos de la lista sean frutas. Para esto, por cada fruta compruebe que:
    1. La fruta sea diccionario.
    2. El diccionario entregado tenga las llaves `nombre`, `tipo`, `color` y `pepas`.

````python
class Jugera(Electrodomestico):
    
    def __init__(self):
        super().__init__()
        self._bandeja = []

    @property
    def bandeja(self):
        return self.bandeja
        
        
    @bandeja.setter
    def bandeja(self, nueva_bandeja):
        if type(nueva_bandeja) is not list:
            raise ValueError("La nueva bandeja debe ser una lista.")

        if len(nueva_bandeja) > 3:
            raise ValueError("No se pueden agregar m√°s de 3 frutas a la bandeja.")
        
        for fruta in nueva_bandeja:
            if not isinstance(fruta, dict):
                raise ValueError("Cada elemento de la nueva bandeja debe ser un diccionario.")
            if not all(key in fruta for key in ["nombre", "tipo", "color", "pepas"]):
                raise ValueError("Cada fruta debe tener las llaves 'nombre', 'tipo', 'color' y 'pepas'.")

        self._bandeja = nueva_bandeja

    def agregar_ingrediente(self, nueva_fruta):

        self.esta_enchufado()  # Verificar si est√° enchufado antes de agregar ingredientes
        if len(self._bandeja) < 3:
            self._bandeja.append(nueva_fruta)
        else:
            raise Exception('Error ‚ùå: No se pueden agregar m√°s de 3 frutas a la bandeja.')
        
    def listar_ingredientes(self):
        if not self._bandeja:
            print("Bandeja vac√≠a")
        else:
            ingredientes = ', '.join([fruta["nombre"] for fruta in self._bandeja])
            print(f"Ingredientes en la bandeja: {ingredientes}.")
    
    def preparar_jugo(self):
        self.esta_enchufado()  # Verificar si est√° enchufado
        
        if not self._bandeja:
            raise Exception('Error ‚ùå: La bandeja no tiene ingredientes.')
        
        frutas_con_pepas = filtrar_por_pepa(self._bandeja, True)
        if frutas_con_pepas:
            print('Alerta ‚ö†Ô∏è: El jugo puede contener restos de pepas.')
        
        nombres_frutas = [fruta["nombre"] for fruta in self._bandeja]
        mensaje = f'Jugo de {", ".join(nombres_frutas)} listo. üèñÔ∏èü•§ Que lo disfrutes!!! ü•§üèñÔ∏è.'
        self._bandeja.clear()  # Vaciar la bandeja
        return mensaje
    
````
