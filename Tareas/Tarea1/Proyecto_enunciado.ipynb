```python
class Profiler():

    # 1. El m√©todo constructor, el cual debe recibir los datos a procesar en formato `Pandas DataFrame`.
    #  Adem√°s, este m√©todo debe generar una carpeta en su directorio de trabajo con el nombre `EDA_fecha`,
    #  donde `fecha` corresponda a la fecha de ejecuci√≥n en formato `DD-MM-YYYY`.

    def __init__(self, dataframe_in):

        self.directorio_actual = os.getcwd()

        # Nos aseguramos que sea una dataframe de pandas para crear la carpeta
        if isinstance(dataframe_in, pd.DataFrame):

            self.dataframe = dataframe_in
            # Obtenemos los datos de hoy y agregamos la parte de 'EDA' para tener el str completo
            now = date.today()
            self.carpeta_nombre = 'EDA_' + now.strftime("%d-%m-%Y")


            # Creamos un condicional para crear la carpeta solo cuando existe
            if not os.path.exists(self.carpeta_nombre):
                os.mkdir(self.carpeta_nombre)
                print(f'Se ha creado la carpeta: {self.carpeta_nombre}')

            # Si no printear un error donde la carpeta ya existe    
            else:
                raise TypeError(
                "Ya se creo una carpeta con este nombre, por favor eliminar la carpeta"
            )  
                     
        # Si no, tirar error
        else:
            raise TypeError(
                "Debes ingresar datos en formato de Pandas DataFrame"
            )        
        
    
    #2. El m√©todo `summarize`, el cual debe caracterizar las variables del Dataset. Como m√≠nimo, se espera que su m√©todo pueda:
    #- Implementar una funcionalidad para filtrar y aplicar este m√©todo a una o m√°s variables de inter√©s. (listo)
    #- Reportar el tipo de variable (ok)
    #- Reportar el n√∫mero y/o porcentaje de valores √∫nicos de la variable (OK)
    #- Reportar el n√∫mero y/o porcentaje de valores nulos (OK)
    #- Si la variables es num√©rica:
    #    - Reportar el n√∫mero y/o porcentaje de valores cero, negativos y outliers
    #    - Reportar estad√≠stica descriptiva como el valor m√≠nimo, m√°ximo, promedio y los percentiles 25, 50, 75 y 100
    # - Levantar una alerta en caso de encontrar alguna anomal√≠a fuera de lo com√∫n (el criterio debe ser ajustable por el usuario)
    #- Guardar sus resultados en el directorio `EDA_fecha/summary.txt`. El archivo debe separar de forma clara y ordenada los resultados de cada punto.
    
    def summarize(self,anomalia = 25, variables=None):

        def calcular_outliers(columna):
            Q1 = columna.quantile(0.25)
            Q3 = columna.quantile(0.75)
            IQR = Q3 - Q1
            inferior = Q1 - 1.5 * IQR
            superior = Q3 + 1.5 * IQR
            outliers = (columna < inferior) | (columna > superior)
            return outliers.sum()

        ruta = os.path.join(self.carpeta_nombre, 'summary.txt')
        with open(ruta, 'w') as archivo_summary:
            for column in self.dataframe.columns:
                if variables and column not in variables:
                    continue  # Salta la columna si no est√° en la lista de inter√©s
                variable_summary = {}
                variable_summary['Variable'] = column
                variable_summary['Tipo'] = str(self.dataframe[column].dtype)
                variable_summary['Valores Unicos'] = self.dataframe[column].nunique()
                variable_summary['Valores Nulos'] = self.dataframe[column].isnull().sum()
                variable_summary['Porcentaje Nulos'] = (variable_summary['Valores Nulos'] / len(self.dataframe)) * 100

                if self.dataframe[column].dtype in ['int64', 'float64']:
                    variable_summary['Valores Cero'] = (self.dataframe[column] == 0).sum()
                    variable_summary['Valores Negativos'] = (self.dataframe[column] < 0).sum()
                    variable_summary['Valores Outliers'] = calcular_outliers(self.dataframe[column])
                    variable_summary['Minimo'] = self.dataframe[column].min()
                    variable_summary['Maximo'] = self.dataframe[column].max()
                    variable_summary['Promedio'] = self.dataframe[column].mean()
                    variable_summary['Percentil 25'] = self.dataframe[column].quantile(0.25)
                    variable_summary['Percentil 50'] = self.dataframe[column].median()
                    variable_summary['Percentil 75'] = self.dataframe[column].quantile(0.75)

                # Alerta por anomal√≠a (por ejemplo, m√°s del 10% de valores nulos)
                if variable_summary['Porcentaje Nulos'] > anomalia:
                    archivo_summary.write(f"Alerta: La variable '{variable_summary['Variable']}' tiene mas del {anomalia}% de valores nulos.\n")

            
                # Escribir el resumen en el archivo
                archivo_summary.write(f"Variable: {variable_summary['Variable']}\n")
                archivo_summary.write(f"Tipo: {variable_summary['Tipo']}\n")
                archivo_summary.write(f"Valores Unicos: {variable_summary['Valores Unicos']}\n")
                archivo_summary.write(f"Valores Nulos: {variable_summary['Valores Nulos']}\n")
                archivo_summary.write(f"Porcentaje Nulos: {variable_summary['Porcentaje Nulos']:.2f}%\n")
                # Escribir m√°s estad√≠sticas si es necesario
                archivo_summary.write(f"Valores Cero: {variable_summary.get('Valores Cero', 0)}\n")
                archivo_summary.write(f"Valores Negativos: {variable_summary.get('Valores Negativos', 0)}\n")
                archivo_summary.write(f"Valores Outliers: {variable_summary.get('Valores Outliers', 0)}\n")
                archivo_summary.write(f"Minimo: {variable_summary.get('Minimo', 0)}\n")
                archivo_summary.write(f"Maximo: {variable_summary.get('Maximo', 0)}\n")
                archivo_summary.write(f"Promedio: {variable_summary.get('Promedio', 0)}\n")
                archivo_summary.write(f"Percentil 25: {variable_summary.get('Percentil 25', 0)}\n")
                archivo_summary.write(f"Percentil 50: {variable_summary.get('Percentil 50', 0)}\n")
                archivo_summary.write(f"Percentil 75: {variable_summary.get('Percentil 75', 0)}\n")
                # L√≠nea separadora
                archivo_summary.write("-" * 50 + "\n")

        print(f"Resumen de variables guardado en {ruta}")

#3 El m√©todo `plot_vars`, el cual debe graficar la distribuci√≥n e interraciones de las variables del Dataset. Como m√≠nimo, se espera que su m√©todo pueda:#
    #- Crear la carpeta `EDA_fecha/plots`
    #- Implementar una funcionalidad para filtrar y aplicar este m√©todo a una o m√°s variables de inter√©s.
    #- Para las variables num√©ricas:
    #    - Genere un gr√°fico de distribuci√≥n de densidad
    #    - Grafique la correlaci√≥n entre las variables
    #- Para las variables categ√≥ricas:
    #    - Genere un histograma de las top N categor√≠as (N debe ser un par√°metro ajustable)
    #    - Grafique el coeficiente V de Cramer entre las variables
    #- Guardar cada gr√°fico generado en la carpeta `EDA_fecha/plots` en formato `.pdf` y bajo el naming `variable.pdf`, donde `variable` es el nombre de la variable de inter√©s
            


    def plot_vars(self, variables=None, top_categorias=5):
        # Crea la carpeta para los gr√°ficos si no existe
        carpeta_plots = os.path.join(self.carpeta_nombre, 'plots')
        if not os.path.exists(carpeta_plots):
            os.mkdir(carpeta_plots)

        # Filtra las variables si se proporcionan
        if variables:
            variables = [var for var in variables if var in self.dataframe.columns]
        else:
            variables = self.dataframe.columns

        for var in variables:
            if var in self.dataframe.select_dtypes(include=['int64', 'float64']).columns:
                # Gr√°fico de distribuci√≥n de densidad para variables num√©ricas
                plt.figure(figsize=(8, 6))
                sns.histplot(self.dataframe[var], kde=True)
                plt.xlabel(var)
                plt.ylabel('Densidad')
                plt.title(f'Distribuci√≥n de {var}')
                plt.savefig(os.path.join(carpeta_plots, f'{var}.pdf'))
                plt.close()

                # Graficar la correlaci√≥n entre las variables num√©ricas
                plt.figure(figsize=(10, 8))
                sns.heatmap(self.dataframe.corr(), annot=True, cmap='coolwarm')
                plt.title('Correlaci√≥n entre variables num√©ricas')
                plt.savefig(os.path.join(carpeta_plots, 'correlation_matrix.pdf'))
                plt.close()

            elif var in self.dataframe.select_dtypes(include=['object']).columns:
                # Histograma de las top N categor√≠as para variables categ√≥ricas
                categorias_top = self.dataframe[var].value_counts().head(top_categorias)
                plt.figure(figsize=(8, 6))
                sns.barplot(x=categorias_top.index, y=categorias_top.values)
                plt.xlabel(var)
                plt.ylabel('Frecuencia')
                plt.title(f'Principales {top_categorias} categor√≠as de {var}')
                plt.savefig(os.path.join(carpeta_plots, f'{var}_categorias_top.pdf'))
                plt.close()

                categorical_vars = self.dataframe.select_dtypes(include=['object']).columns

                for var1, var2 in itertools.combinations(categorical_vars, 2):
                # Verifica si hay suficientes combinaciones de valores √∫nicos
                    if len(set(self.dataframe[var1].unique()) & set(self.dataframe[var2].unique())) >= 2:
                    # Calcula el coeficiente V de Cramer
                        contingency_table = pd.crosstab(self.dataframe[var1], self.dataframe[var2])
                        chi2, _, _, _ = chi2_contingency(contingency_table)
                        n = contingency_table.sum().sum()
                        phi2 = chi2 / n
                        r, k = contingency_table.shape
                        phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))
                        rcorr = r - ((r-1)**2)/(n-1)
                        kcorr = k - ((k-1)**2)/(n-1)
                        cramers_v = np.sqrt(phi2corr / min((kcorr-1), (rcorr-1)))

                        # Crea el gr√°fico del coeficiente V de Cramer
                        plt.figure(figsize=(8, 6))
                        sns.heatmap([[cramers_v]], annot=True, cmap='coolwarm', cbar=False, square=True, fmt=".2f")
                        plt.xticks([])
                        plt.yticks([])
                        plt.title(f'Coeficiente V de Cramer entre {var1} y {var2}')
                        plt.savefig(os.path.join(carpeta_plots, f'{var1}_{var2}_cramer.pdf'))
                        plt.close()
                    else:
                        print(f"No hay suficientes combinaciones de valores √∫nicos entre {var1} y {var2}.")

#4 El m√©todo `clean_data`, el cual debe limpiar los datos para que luego puedan ser procesados. Como m√≠nimo, se espera que su m√©todo pueda:
#    - Crear la carpeta `EDA_fecha/clean_data`
#   - Implementar una funcionalidad para filtrar y aplicar este m√©todo a una o m√°s variables de inter√©s.
#    - Drop de valores duplicados
#    - Implementar como m√≠nimo 2 t√©cnicas para tratar los valores nulos, como:
#        - Drop de valores nulos
#        - Imputar valores nulos con alguna t√©cnica de imputaci√≥n
#        - Funcionalidad para escoger entre una t√©cnica y la otra.
#    - Una de las columnas del dataframe presenta datos *no at√≥micos*. Separe dicha columna en las columnas que la compongan.
#        - Hint: ¬øQu√© caracteres permiten separar una columna de otra?
#        - Para las pruebas con el dataset nuevo, puede esperar que exista al menos una columna con este tipo de problema. Asuma que los separadores ser√°n los mismos, aunque el n√∫mero de columnas a separar puede ser distinto.
#    - Deber√≠an usar `FunctionTransformer`.
#    - Guardar los datos procesados en formato `.csv` en el path `EDA_fecha/clean_data/data.csv`

    def clean_data(self):

        # Creamos el directorio donde crear la nueva carpeta
        now = date.today()
        ruta = 'EDA_' + now.strftime("%d-%m-%Y") 
        nombre_carpeta = "clean_data"
        ruta_completa = os.path.join(ruta, nombre_carpeta)

        # Nos aseguramos de que no existe
        try:
            os.makedirs(ruta_completa)
            print(f'Se ha creado la carpeta "{ruta_completa}" con √©xito.')
        except FileExistsError:
            print(f'La carpeta "{ruta_completa}" ya existe.')
        
        # Tomamos los datos de la funcion anterior

        df = self.dataframe
        df2 = df.copy()
        df3 = df.copy()
        #####
        # Solucionamos el problema de la tabla con datos no atomicos
        # Iterar a trav√©s de las columnas del DataFrame
        for columna in df3.columns:
            if '-' in columna:
                # Dividir la columna en m√∫ltiples columnas usando el gui√≥n como separador
                nuevas_columnas = columna.split('-')
            
                for nueva_columna in nuevas_columnas:
                    df2[nueva_columna] = None

        # Eliminar las columnas originales con guiones
        df2.drop(df3.columns, axis=1, inplace=True)

        df3 = df3.astype(str)

        # Iterar a trav√©s de las columnas del DataFrame
        for columna in df3.columns:
            # Separar cada fila en funci√≥n de los valores entre par√©ntesis
            for index, valor in enumerate(df3[columna]):
                if isinstance(valor, str) and( ("*" and "?") or ("(" and "?") or (":" and "?") ) in valor:    # vemos cada caso donde hay datos mal imputados
                    valores_entre_parentesis = re.findall(r'\d+\.\d+', valor)
                    for i, val in enumerate(valores_entre_parentesis):
                        nueva_columna = f"{columna}_{i+1}"
                        df2.at[index, nueva_columna] = float(val)
        
      
  
        
        for columna in df2.columns:
            if df2[columna].isnull().all():
                # Eliminar la columna si todos los valores son nulos
                df2 = df2.drop(columna, axis=1)

        # Reemplaza los nombres de las columnas y elimina "_ y n√∫meros"
        df2.columns = df2.columns.str.replace(r'_', '') 
        df2.columns = df2.columns.str.replace(r'1', '') 
        df2.columns = df2.columns.str.replace(r'2', '') 
        df2.columns = df2.columns.str.replace(r'3', '') 
        df2.columns = df2.columns.str.replace(r'4', '') 
        df2.columns = df2.columns.str.replace(r'5', '')


        columnas_nuevas = df2.columns
        columnas_nuevas = list(set(columnas_nuevas))

        # Lista transformada
        lista_transformada = [subcadena for elemento in columnas_nuevas for subcadena in elemento.split('-')]

        df2.columns = lista_transformada

        # Ahora concatenamos la tabla con las nuevas columnas atomicas y la tabla sin las columnas atomicas
        columnas_a_eliminar = [col for col in df3.columns if '-' in col]
        df3 = df3.drop(columns=columnas_a_eliminar)

        df2_3 = pd.concat([ df3, df2], axis=1)

        # TERMINAMOS CON ARREGLAR LA no ATOMICIDAD
        #####

        ######
        # Dropeamos los valores duplicados

        df2_3 = df2_3.drop_duplicates()

    

        # Preguntar a quien aplicar el clean
        for indice, i in enumerate(df2_3.columns, start=0):
            print(f"{indice}: {i}")

        # Preguntamos al usuario que desea limpiar
        seleccion = input("Defina que variables quiere limpiar, si quiere aplicar a todas la variables escriba ""all"", sino escriba como lista las variables de inter√©s de la lista printeada, por ejemplo si desea limpiar las variables ""year"" y ""season"" tiene que ingresar ""[4,5]")
        
        # Si escribe all elige la opcion "0"
        if seleccion == "all":

            seleccion2 = input("Seleccione el m√©todo (ponga solo el numero). 1:Drop valores nulos, 2:Imputacion por moda")
            
            if seleccion2 == '1':
                df2_3.fillna(0, inplace=True)
                #####
                # Especifica la ruta del archivo CSV
                now = date.today()
                ruta = 'EDA_' + now.strftime("%d-%m-%Y")
                ruta_csv = ruta+'/clean_data/data.csv'
                # Guarda el DataFrame en el archivo CSV
                df2_3.to_csv(ruta_csv, index=False)
                print(f"Los datos se han guardado en {ruta_csv}")
                ######
                return
         

            elif seleccion2 == '2':
                mode = df2_3.mode().iloc[0]
                df2_3.fillna(mode, inplace=True)
                                #####
                # Especifica la ruta del archivo CSV
                now = date.today()
                ruta = 'EDA_' + now.strftime("%d-%m-%Y")
                ruta_csv = ruta+'/clean_data/data.csv'
                # Guarda el DataFrame en el archivo CSV
                df2_3.to_csv(ruta_csv, index=False)
                print(f"Los datos se han guardado en {ruta_csv}")
                ######
                return
                
            
    
        # Si escribe algo que no sea all, hay que verificar que sea una lista
        else: 
            try:
                lista = eval(seleccion)
                if isinstance(lista, list):
                    print("La variable 'lista' es una lista.")
                    seleccion3 = input("Que tecnica desea aplicar? (ponga solo el numero). 1:Drop valores nulos, 2:Imputaci√≥n seg√∫n media( si selecciono variables categoricas esto no funcionara), 3:Imputacion por moda")
                
                if seleccion3 == '1':
                    # Rellenar valores faltantes en las columnas seleccionadas con cero
                    df2_3.iloc[:, lista] = df2_3.iloc[:, lista].fillna(0)
                                    #####
                    # Especifica la ruta del archivo CSV
                    now = date.today()
                    ruta = 'EDA_' + now.strftime("%d-%m-%Y")
                    ruta_csv = ruta+'/clean_data/data.csv'
                    # Guarda el DataFrame en el archivo CSV
                    df2_3.to_csv(ruta_csv, index=False)
                    print(f"Los datos se han guardado en {ruta_csv}")
                    ###### 
                    return 
        
                elif seleccion3 == '2':
                    # Sacamos la media 
                    imputer = SimpleImputer(strategy='mean')
                    df2_3.iloc[:, lista] = imputer.fit_transform(df2_3.iloc[:, lista].values)
                    #####
                    # Especifica la ruta del archivo CSV
                    now = date.today()
                    ruta = 'EDA_' + now.strftime("%d-%m-%Y")
                    ruta_csv = ruta+'/clean_data/data.csv'
                    # Guarda el DataFrame en el archivo CSV
                    df2_3.to_csv(ruta_csv, index=False)
                    print(f"Los datos se han guardado en {ruta_csv}")
                    ######
                    return   

                elif seleccion3 == '3':
                    # Calcular la moda de las columnas seleccionadas
                    mode = df2_3.iloc[:, lista].mode().iloc[0]
                    df2_3.iloc[:, lista] = df2_3.iloc[:, lista].fillna(mode)
                    #####
                    # Especifica la ruta del archivo CSV
                    
                    now = date.today()
                    ruta = 'EDA_' + now.strftime("%d-%m-%Y")
                    ruta_csv = ruta+'/clean_data/data.csv'
                    # Guarda el DataFrame en el archivo CSV
                    df2_3.to_csv(ruta_csv, index=False)
                    print(f"Los datos se han guardado en {ruta_csv}")
                    ######
                    return 


                else:
                    print("La variable 'lista' no es una lista.")
            except SyntaxError:
                print("La cadena no es una lista v√°lida.") 


#5. El m√©todo `scale`, el cual debe preparar adecuadamente los datos para luego ser consumidos por alg√∫n tipo de algoritmo. Como m√≠nimo, se espera que su m√©todo pueda:
#    - Crear la carpeta `EDA_fecha/scale`
#    - Procesar de forma adecuada los datos num√©ricos y categ√≥ricos:
#        - Su m√©todo debe recibir las t√©cnicas de escalamiento como argumento de entrada (utilizar solo t√©cnicas compatibles con el framework de `sklearn`)
#        - Para los atributos num√©ricos, se transforme los datos con un escalador logar√≠tmico y un `MinMaxScaler`
#        - Asuma que no existen datos ordinales en su dataset
#    - Guardar todo este procesamiento en un `ColumnTransformer`.
#    - Guardar los datos limpios y transformados en formato `.csv` en el path `EDA_fecha/process/scaled_features.csv`

    def scale(self, scale_numerico=[MinMaxScaler(),FunctionTransformer(func=np.log1p, inverse_func=np.expm1, validate=False)]):

        # Carga el archivo CSV con datos limpios

        now = date.today()
        ruta = 'EDA_' + now.strftime("%d-%m-%Y") 
        nombre_carpeta2 = "clean_data/data.csv"
        ruta_completa2 = os.path.join(ruta, nombre_carpeta2)

        # Nos aseguramos de que no existe
        try:
            os.makedirs(ruta_completa2)
            print(f'Se ha creado la carpeta "{ruta_completa2}" con √©xito.')
        except FileExistsError:
            print(f'La carpeta "{ruta_completa2}" ya existe.')


        data_limpiada = pd.read_csv(ruta_completa2) 

        #data_limpiada = pd.read_csv("D:/Universidad/Laboratorio de Programacion/MDS7202-1-Primavera-2023/Tareas/Tarea1/EDA_06-11-2023/clean_data/data.csv")

        # Crea la carpeta para los datos procesados si no existe
        carpeta_procesados = os.path.join(self.carpeta_nombre, 'process')
        if not os.path.exists(carpeta_procesados):
            os.mkdir(carpeta_procesados)

        # Obt√©n las columnas de las variables num√©ricas y categ√≥ricas
        columnas_numericas = data_limpiada.select_dtypes(include=['int64', 'float64']).columns.tolist()
        columnas_categoricas = ['Sex', 'NOC', 'Year', 'Season', 'City', 'Sport', 'Medal']

        # Define los transformadores para las variables num√©ricas y categ√≥ricas
        
        numeric_transformer = Pipeline(steps=[
                ('log_transform', scale_numerico[1]),
                ('minmax_scale', scale_numerico[0])
            ])

        categorical_transformer = OneHotEncoder(drop='first', sparse=False)

        # Crea el ColumnTransformer final
        
        preprocessor = ColumnTransformer(
            transformers=[
                ('numeric', numeric_transformer, columnas_numericas),
                ('categorical', categorical_transformer, columnas_categoricas)
            ])
        
        # Aplica el preprocesador a los datos
        processed_data = preprocessor.fit_transform(data_limpiada)

        # Despu√©s de aplicar el preprocesador, obt√©n las columnas transformadas
        columnas_transformadas = (
            
            
            ['min_max_' + col for col in columnas_numericas] +
            preprocessor.named_transformers_['categorical'].get_feature_names_out(columnas_categoricas).tolist()
        )

        # Crea un DataFrame con los datos transformados y las columnas correspondientes
        processed_df = pd.DataFrame(processed_data, columns=columnas_transformadas) 

        # Guarda los datos transformados en un archivo CSV
        scaled_features_path = os.path.join(carpeta_procesados, 'scaled_features.csv')
        processed_df.to_csv(scaled_features_path, index=False)
        print(f"Datos transformados guardados en {scaled_features_path}")

#6. El m√©todo `make_clusters`, el cual debe generar clusters de los datos usando alg√∫n algoritmo de clusterizaci√≥n. Como m√≠nimo, se espera que su m√©todo pueda:
#    - Crear la carpeta `EDA_fecha/clusters`
#    - Generar un estudio del codo donde se√±ale la cantidad de clusters optimos para el desarrollo.
#    - Su m√©todo debe recibir el algoritmo de clustering como argumento de entrada (utilizar solo algoritmos compatibles con el framework de `sklearn`).
#    - No olvide pre procesar adecuadamente los datos antes de implementar la t√©cnica de clustering. 
#    - En este punto es espera que generen un `Pipeline` de sklearn. Adem√°s, su m√©todo deber√≠a usar lo construido en los puntos 4 y 5. 
#    - Su m√©todo debe ser capaz de funcionar a partir de datos crudos (se descontar√° puntaje de lo contrario).
#    - Una vez generado los clusters, proyecte los datos a 2 dimensiones usando su t√©cnica de reducci√≥n de dimensionalidad favorita y grafique los resultados coloreando por cluster.
#    - Guardar los datos con su respectivo cluster en formato `.csv` en el path `EDA_fecha/clusters/data_clusters.csv`. Guarde tambi√©n los gr√°ficos generados en el mismo path.

    def make_clusters(self, clustering_algorithm=KMeans(random_state=42, n_init=10)):

        now = date.today()
        ruta = 'EDA_' + now.strftime("%d-%m-%Y") 
        nombre_carpeta3 = "process/scaled_features.csv"
        ruta_completa3 = os.path.join(ruta, nombre_carpeta3)

        # Nos aseguramos de que no existe
        try:
            os.makedirs(ruta_completa3)
            print(f'Se ha creado la carpeta "{ruta_completa3}" con √©xito.')
        except FileExistsError:
            print(f'La carpeta "{ruta_completa3}" ya existe.')

        data_cluster = pd.read_csv(ruta_completa3) 

        # Crea la carpeta para los clusters si no existe
        carpeta_clusters = os.path.join(self.carpeta_nombre, 'clusters')
        if not os.path.exists(carpeta_clusters):
            os.mkdir(carpeta_clusters)

        # Combina preprocesamiento y algoritmo de clustering en un Pipeline
        pipeline = Pipeline([
            
            ('clustering', clustering_algorithm)
        ])

        # Encuentra la cantidad √≥ptima de clusters usando el m√©todo del codo

        # Encuentra el n√∫mero √≥ptimo de clusters usando la derivada segunda del m√©todo del codo
        max_clusters = 20

        second_derivative = []
        for k in range(1, max_clusters + 1):
            pipeline.set_params(clustering__n_clusters=k)
            pipeline.fit(data_cluster)
            inertia = pipeline.named_steps['clustering'].inertia_
            second_derivative.append(inertia)
        
        second_derivative = np.gradient(np.gradient(second_derivative))  # Segunda derivada
        optimal_num_clusters = np.argmax(second_derivative) + 1  # Agrega 1 porque el √≠ndice comienza desde 0

        # Aplica clustering con el n√∫mero √≥ptimo de clusters
        pipeline.set_params(clustering__n_clusters=optimal_num_clusters)
        clusters = pipeline.fit_predict(data_cluster)

        # Reduce la dimensionalidad a 2 dimensiones para la visualizaci√≥n
        pca = PCA(n_components=2)
        reduced_features = pca.fit_transform(data_cluster)
        reduced_cluster_centers = pca.transform(pipeline.named_steps['clustering'].cluster_centers_)

        # Guarda los datos con sus respectivos clusters en formato .csv
        clustered_data = pd.DataFrame({'Cluster': clusters})
        clustered_data.to_csv(os.path.join(carpeta_clusters, 'data_clusters.csv'), index=False)

        # Grafica los resultados coloreando por cluster
        plt.figure(figsize=(8, 6))
        plt.scatter(reduced_features[:, 0], reduced_features[:, 1], c=clusters, cmap='viridis', alpha=0.6)
        plt.scatter(reduced_cluster_centers[:, 0], reduced_cluster_centers[:, 1], marker='x', s=200, linewidths=3, color='red')
        plt.xlabel('Dimensi√≥n 1')
        plt.ylabel('Dimensi√≥n 2')
        plt.title('Proyecci√≥n de Datos en 2 Dimensiones Coloreada por Cluster')
        plt.savefig(os.path.join(carpeta_clusters, 'cluster_visualization.png'))
        plt.close()

        print("Clusters generados y guardados correctamente.")

#7. El m√©todo `detect_anomalies`, el cual debe detectar anomal√≠as en los datos. Como m√≠nimo, se espera que su m√©todo pueda:

#    - Crear la carpeta `EDA_fecha/anomalies`
#    - Implementar alguna t√©cnica de detecci√≥n de anomal√≠as.
#    - Al igual que el punto anterior, su m√©todo debe considerar los siguientes puntos:
#        - No olvide pre procesar de forma adecuada los datos antes de implementar la t√©cnica de detecci√≥n de anomal√≠a. 
#        - En este punto es espera que generen un `Pipeline` de sklearn. Adem√°s, su m√©todo deber√≠a usar lo construido en los puntos 4 y 5. 
#        - Su m√©todo debe ser capaz de funcionar a partir de datos crudos (se descontar√° puntaje de lo contrario).
#        - Su m√©todo debe recibir el algoritmo como argumento de entrada
#        - Una vez generado las etiquetas, proyecte los datos a 2 dimensiones y grafique los resultados coloreando por las etiquetas predichas por el detector de anomal√≠as
#    - Guardar los datos con su respectiva etiqueta en formato `.csv` en el path `EDA_fecha/anomalies/data_anomalies.csv`. Guarde tambi√©n los gr√°ficos generados en el mismo path.


    def detect_anomalies(self, anomaly_detector=GaussianMixture(random_state=42, n_components=2)):
        # Crea la carpeta para las anomal√≠as si no existe
        carpeta_anomalies = os.path.join(self.carpeta_nombre, 'anomalies')
        if not os.path.exists(carpeta_anomalies):
            os.mkdir(carpeta_anomalies)

        # Preprocesamiento de datos
        preprocessor = joblib.load(os.path.join(self.carpeta_nombre, 'process', 'preprocessor.pkl'))
        processed_data = preprocessor.transform(self.dataframe)

        # Entrenar el detector de anomal√≠as
        anomaly_detector.fit(processed_data)
        anomaly_labels = anomaly_detector.predict(processed_data)

        # Guardar las etiquetas de anomal√≠a en el archivo CSV
        output_csv_path = os.path.join(carpeta_anomalies, 'data_anomalies.csv')
        processed_df = pd.DataFrame(processed_data, columns=self.dataframe.columns)
        processed_df['Anomaly_Label'] = anomaly_labels
        processed_df.to_csv(output_csv_path, index=False)
        print(f"Datos con etiquetas de anomal√≠a guardados en {output_csv_path}")

        # Proyectar los datos a 2 dimensiones usando PCA
        pca = PCA(n_components=2, random_state=42)
        projected_data = pca.fit_transform(processed_data)

        # Graficar los datos coloreando por las etiquetas de anomal√≠a
        plt.figure(figsize=(8, 6))
        plt.scatter(projected_data[:, 0], projected_data[:, 1], c=anomaly_labels, cmap='coolwarm')
        plt.xlabel('Principal Component 1')
        plt.ylabel('Principal Component 2')
        plt.title('Proyecci√≥n de Datos con Etiquetas de Anomal√≠a')
        plt.colorbar(label='Anomaly Label')
        plt.savefig(os.path.join(carpeta_anomalies, 'anomalies_plot.png'))
        plt.close()
        print("Gr√°fico de anomal√≠as guardado.")

    #8 METODOOOOOO
    def clearGarbage(self):




        try:
            for carpeta in os.listdir(self.directorioactual):
                if carpeta.startswith("EDA") and os.path.isdir(carpeta):
                    ruta_carpeta = os.path.join(self.directorio_actual, carpeta)
                    # Eliminar los archivos y subdirectorios dentro de la carpeta
                    for root, dirs, files in os.walk(ruta_carpeta, topdown=False):
                        for name in files:
                            os.remove(os.path.join(root, name))
                        for name in dirs:
                            os.rmdir(os.path.join(root, name))
                    # Eliminar la carpeta principal
                    os.rmdir(ruta_carpeta)
                    print(f"Se ha eliminado la carpeta: {carpeta}")
        except Exception as e:
            print(f"Error al eliminar carpetas: {e}")
```
